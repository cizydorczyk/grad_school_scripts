```r
pdf("E4_length_vs_coverage_all_isolates.pdf", width=50, height=40)
par(mfrow=c(8,10))

for (file in files){
    name = basename(file)
    print(name)
    data = read.table(file, header=TRUE, sep='\t')
    #head(data1$Outline)
    #head(data1$Color)
    data1 = transform(data, Color=as.character(Color), Outline=as.character(Outline))
    plot(log10(data1$Length), log10(data1$Coverage), pch=21, cex=0.8, col=data1$Outline, bg=data1$Color, xlab="Contig Length (log)", ylab="Contig Coverage (log)", las=1, main=name)
    
}

dev.off()
```
